# Exploratory Data Analysis (EDA)

## 1. Introduction

This notebook demonstrates how to perform Exploratory Data Analysis (EDA) on a dataset using Python. The steps include:

1. Loading the dataset.
2. Understanding the data structure and types.
3. Summarizing statistics of numerical and categorical features.
4. Identifying and visualizing correlations between features.
5. Detecting and handling outliers if necessary.
6. Drawing meaningful insights and conclusions from the data.

## 6. Identify and Visualize Correlations Between Features

Weâ€™ll use visualizations to understand the relationships between different features.

### 6.1 Histogram

We'll create a histogram to understand the distribution of a numerical feature like Male Risk Development Percentage.

```python
df['Male Risk Development Percentage'].hist()
plt.xlabel('Male Risk Development Percentage')
plt.ylabel('Frequency')
plt.title('Distribution of Male Risk Development Percentage')
plt.show()
```

### 6.2 Bar Plot

We'll create a bar plot to visualize the count of different categories within Cancer Type.

```python
df['Cancer Type'].value_counts().plot(kind='bar')
plt.xlabel('Cancer Type')
plt.ylabel('Count')
plt.title('Count of Each Cancer Type')
plt.show()
```

### 6.3 Scatter Plot

We'll create a scatter plot to explore the relationship between Male Risk Development Percentage and Male Risk Dying Percentage.



```python
df.plot(kind='scatter', x='Male Risk Development Percentage', y='Male Risk Dying Percentage')
plt.xlabel('Male Risk Development Percentage')
plt.ylabel('Male Risk Dying Percentage')
plt.title('Male Risk Development Percentage vs. Male Risk Dying Percentage')
plt.show()
```

###  6.4 Seaborn Heatmap

Weâ€™ll create a heatmap to visualize the correlation between different numerical variables.

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## 7. Detect and Handle Outliers

Outliers can significantly impact the results of your analysis, so itâ€™s important to identify and handle them if necessary.

```python
Q1 = df['Male Risk Development Percentage'].quantile(0.25)
Q3 = df['Male Risk Development Percentage'].quantile(0.75)
IQR = Q3 - Q1

df_outliers_removed = df[~((df['Male Risk Development Percentage'] < (Q1 - 1.5 * IQR)) | (df['Male Risk Development Percentage'] > (Q3 + 1.5 * IQR)))]
```
