# Data Cleaning and Visualization

## 1. Introduction
This notebook demonstrates how to clean and visualize a dataset using Python. The steps include:

1. Loading the dataset.
2. Handling missing values.
3. Removing duplicates.
4. Transforming data.
5. Creating visualizations.

## 2. Import Libraries
First, we need to import the necessary libraries.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

## 3. Load Dataset
Next, we'll load the dataset from the specified path.


```python
df = pd.read_csv(r'C:\Users\Luqman\Desktop\Neevas DataScience Internship\Task 1\languages_dataset.csv')


df.head()
```

## 4. Data Cleaning
### 4.1 Handle Missing Values
First, we'll handle any missing values in the dataset.


```python
df.dropna(inplace=True)
```

### 4.2 Remove Duplicates
Next, we'll remove any duplicate rows to ensure data integrity.


```python
df.drop_duplicates(inplace=True)
```

### 4.3 Data Transformation
We'll perform necessary data transformations

```python
df['Dummy_numeric'] = np.random.randint(1, 100, df.shape[0])
```

## 5. Data Visualization
Now, we'll create various plots to explore the data and identify patterns or trends.
### 5.1 Histogram
We'll start with a histogram to show the distribution of ages.


```python
# Histogram
df['Speakers'].hist()
plt.xlabel('Number of Speakers')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Speakers')
plt.show()
```

### 5.2 Bar Plot
Next, we'll create a bar plot to show the count of different Regions.

```python
# Bar plot
df['Region'].value_counts().plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Count of Languages by Region')
plt.show()
```

### 5.3 Scatter Plot
We'll create a scatter plot to show the relationship between `Speakers` and `Region`.


```python
df['Region_numeric'] = df['Region'].astype('category').cat.codes
df.plot(kind='scatter', x='Region_numeric', y='Speakers')
plt.xlabel('Region (encoded)')
plt.ylabel('Number of Speakers')
plt.title('Speakers by Region')
plt.show()
```

### 5.4 Seaborn Heatmap
Finally, we'll create a heatmap to show the correlation between different numeric variables.


```python
df['Dummy_numeric'] = np.random.randint(1, 100, df.shape[0])

sns.heatmap(df[['Speakers', 'Dummy_numeric']].corr(), annot=True)
plt.title('Correlation Matrix')
plt.show()
```
